```python
import numpy as np
import cv2

# read the image 
image  =  cv2.imread('tiger.jpg')
# gray scale conversion
gray_img = cv2.cvtColor(image,
                       cv2.COLOR_BGR2GRAY)
```

```python
gray_img.shape , image.shape
```

```python
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('image',gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Code
- steps to capture a video:

- use cv2.VideoCapture() to get a video capture object for the camera.
- Video is a combination of frame
- when we capture live video, it is like a infinite frames till we stop the video
- set up a infinite wile loop and use the read() method to read the frames using the above created object.
- Use cv2.imshow() method to show the frames in the video
- Breaks the loop when the user clicks a specific key.

```python
# import cv2

cv2.imread() : Image read
cv2.imshow() : Image show
cv2.imread() : Image write
cv2.VideoCapture(0)  # it will open the camera laptop. capture our videos.
```

```python
# import the opencv lib
import cv2

# define the video capture object 
vid = cv2.VideoCapture(0)

while True:
    # Capture the video frame 
    # by frame 
    
    ret, frame = vid.read()
    # ret = boolean value 
    # ret = True, the frame is captured 
    # ret = False, the frame is not captured 
    print(ret)
    # display the resulting frame 
    cv2.imshow('frame', frame)
    # the 'q' button is set as the 
    # quitting button you may use any 
    # desired button of any choice 
    # 1 means one milisec in waitkey
    # 1 and 0xFF its try to identify an 8 digit key 
    # ord('q') : ASCII 8 digits
    # so it will check are you entered any 8 digit key or not 
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
# after the loop release the cap object 
vid.release()
# Destroy all the windows 
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture(0)

while(True):


    # Capture frames in the video 
    ret, frame = cap.read()
    # describe the gtype of the data to be used 
    font = cv2.FONT_HERSHEY_SIMPLEX

    # use putText method() for this 
    # inserting text on the video R=0,G=255,B=255

    cv2.putText(frame,
                'Hiee Pratik',
                (50,50), 
                font, 1, 
                (0,255,255), 
                2, 
                cv2.LINE_4)

    # display the resulting frame 
    cv2.imshow('video', frame)

    # creating q as the quit 
    # button for te video

    if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
